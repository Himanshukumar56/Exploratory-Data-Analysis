```python
ax = df['Year_Introduced'].value_counts()\
.head(10)\
.plot(kind='bar',title='Top 10 Years for Roller Coaster Introductions', figsize=(12,6), color='skyblue')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Number of Coasters')
```
